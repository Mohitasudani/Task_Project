```python
# Importing libraries
import urllib.request, urllib.parse, urllib.error
import json
import ssl
import pandas as pd
import urllib.request, urllib.parse, urllib.error
import oauth
# importing hidden.py for keys
import hidden

# Using oauth for security. Also, one should have oauth installed or have outh.py attached on my github repo 
def augment(url, parameters):
    secrets = hidden.oauth()
    consumer = oauth.OAuthConsumer(secrets['consumer_key'],
                                   secrets['consumer_secret'])
    token = oauth.OAuthToken(secrets['token_key'], secrets['token_secret'])

    oauth_request = oauth.OAuthRequest.from_consumer_and_token(consumer,
                    token=token, http_method='GET', http_url=url,
                    parameters=parameters)
    oauth_request.sign_request(oauth.OAuthSignatureMethod_HMAC_SHA1(),
                               consumer, token)
    return oauth_request.to_url()


# Twitter api url for data retrieval 
TWITTER_URL = 'https://api.twitter.com/1.1/statuses/user_timeline.json'

# For Ignoring SSL certificate errors
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE


while True:
    print('')
    # input for twitter account handle and connection through the api with the help of twitter url
    account = input('Enter Twitter Account:')
    if (len(account) < 1): break
    url = augment(TWITTER_URL,{'screen_name': account})
    connection = urllib.request.urlopen(url, context=ctx)
    #decoding to utf-8
    data = connection.read().decode()

    # and Loading the date
    js = json.loads(data)

    # retrieving headers to display remaining limit requests
    headers = dict(connection.getheaders())
    print('Remaining--', headers['x-rate-limit-remaining'])

    # main information retrieval for making table
    date_time=[]
    text_tweet=[]
    like=[]
    retweet=[]
    image_cnt=[]
    for u in js:
        cnt=0
        date_time.append(u['created_at'])
        text_tweet.append(u['text'])
        like.append(u['favorite_count'])
        retweet.append(u['retweet_count'])
        try:
            for i in u['entities']['media']:
                if i['type']=='photo':
                    cnt=cnt+1
                    image_cnt.append(cnt)
        except:
            image_cnt.append(0)
        
    df=pd.DataFrame({'date_time':date_time,'text':text_tweet,'likes':like,'retweets':retweet,'image_count':image_cnt})
    print(df)

```
